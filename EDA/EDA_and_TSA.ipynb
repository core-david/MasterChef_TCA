```python
def plot_distributions(df, max_unique = 20):
    
    for column in df.columns:
        
        if pd.api.types.is_datetime64_any_dtype(df[column]):
            continue  # Skip date columns

        plt.figure(figsize = (8, 4))
        plt.title(f'Distribution of {column}', fontsize = 14)

        if pd.api.types.is_numeric_dtype(df[column]):
            sns.histplot(df[column].dropna(), kde = True, bins = 30)
            
        elif df[column].nunique() <= max_unique:
            sns.countplot(x = column, data = df, order = df[column].value_counts().index)
            plt.xticks(rotation = 45)
            
        else:
            print(f"Skipping '{column}' (too many unique categories: {df[column].nunique()})")
            plt.close()
            continue

        plt.tight_layout()
        plt.show()
```

```python
def sum_isna(df, column):
    
    for value in df[column].dropna().unique():
        na_count = df[df[column] == value].isna().sum().sum()
        amount_rows = len(df[df[column] == value])
        print(f"NAs: {na_count}/{amount_rows}   {round(100 * na_count / amount_rows, 2)}      {value}")
```

Vemos el desglose por cada estatus con valores nulos. Gracias a este analisis, vemos que aquellos registros con valores nulos no acumulan mas del 4% de los registros por cada estatus

Ahora detectamos cuales son las columnas con valores faltantes

```python
def na_summary_by_column(df, column):
    
    values = df[column].dropna().unique()
    summary = pd.DataFrame()

    for value in values:
        filtered = df[df[column] == value]
        na_counts = filtered.isna().sum()
        summary[value] = na_counts
        
    summary = summary.loc[summary.sum(axis = 1) > 0, summary.sum(axis = 0) > 0]

    return summary
```

##### Imputacion de NaNs

Continuemos a aplicar estos cambios acordados para las fechas correspondientes en base a su estatus.

```python
def llenar_nans_fechas(row):
    
    if pd.isna(row['fecha_llegada']):
        
        if pd.isna(row['fecha_salida']):
            row['fecha_salida'] = row['fecha_confirmacion_pickup']
        
        row['fecha_llegada'] = row['fecha_salida'] - row['cantidad_noches']
        
    if pd.isna(row['fecha_salida']):
        row['fecha_salida'] = row['fecha_llegada'] + row['cantidad_noches']
        
    if pd.isna(row['fecha_registro']):
        
        if row['estatus_de_la_reservacion'] == "RESERVACION CANCELADA" or row['estatus_de_la_reservacion'] == "NO SHOW":
            row['fecha_registro'] = row['fecha_reservacion']
            
        elif row['estatus_de_la_reservacion'] == "ROOMING LIST" or row['estatus_de_la_reservacion'] == "RESERVACION O (R)REGISTRO":
            row['fecha_registro'] = row['fecha_confirmacion_pickup']
    
    return row

df = df.apply(llenar_nans_fechas, axis = 1)
```

```python
df[['fecha_llegada', 'fecha_registro', 'fecha_salida']] = df[['fecha_llegada', 'fecha_registro', 'fecha_salida']].astype('int64')
```

```python
fecha_limite = df.fecha_reservacion.min()
print(fecha_limite)

df_fechas_2000 = df[
    (df['fecha_registro'].lt(fecha_limite)) |
    (df['fecha_confirmacion_pickup'].lt(fecha_limite))
]
print(df_fechas_2000.shape)
view(df_fechas_2000)
```

Al visualizar los datos con fechas de 2000 inferimos lo siguiente:
- Registros con estatus 'ROOMING LIST' no tienen ingresos; se asume que los clientes se quedaron en lista de espera y nunca se confirmo la reserva.
- Registros con estatus 'RESERVACION O R(REGISTRO) tienen ingreso; se asume que la reserva si termino llevandose a cabo.

Para los estatus 'ROOMING LIST' 'fecha_confirmacion_pickup' y 'fecha_registro' con fechas previas al 2019 asumiran el mismo valor que 'fecha_llegada'.

Para los estatus 'RESERVACION O (R)REGISTRO' 'fecha_confirmacion_pickup' y 'fecha_registro' asumiran los valores de 'fecha_salida' y 'fecha_llegada' respectivamente.

```python
def llenar_fechas_2000(row):
    
    if row['fecha_registro'] < fecha_limite:
        row['fecha_registro'] = row['fecha_llegada']
        
    if row['fecha_confirmacion_pickup'] < fecha_limite:
        
        if row['estatus_de_la_reservacion'] == "ROOMING LIST":
            row['fecha_confirmacion_pickup'] = row['fecha_llegada']
            
        elif row['estatus_de_la_reservacion'] == "RESERVACION O (R)REGISTRO":
            row['fecha_confirmacion_pickup'] = row['fecha_salida']
    
    return row

del df_fechas_2000
df = df.apply(llenar_fechas_2000, axis=1)
```

Para estos casos particulares parece haber un error. Reemplazaremos el valor de 'fecha_confirmacion_pickup' por 'fecha_registro' o 'fecha_salida' dependiendo de si estan dentro del rango de tiempo.

```python
condicion = (
    (df['fecha_confirmacion_pickup'] > fecha_limite) |
    (df['fecha_registro'] > fecha_limite)
)

def corregir_fechas(row):
    
    if row['fecha_registro'] > fecha_limite:
        row['fecha_registro'] = row['fecha_salida']
    
    if row['fecha_confirmacion_pickup'] > fecha_limite:
        row['fecha_confirmacion_pickup'] = row['fecha_registro']
    
    return row

df.loc[condicion] = df.loc[condicion].apply(corregir_fechas, axis = 1)
```

Analizemos nuevamente las series de tiempo.

```python
df_reservacion = df.set_index('fecha_reservacion')
df_llegada = df.set_index('fecha_llegada')
df_registro = df.set_index('fecha_registro')
df_salida = df.set_index('fecha_salida')
df_conf_pickup = df.set_index('fecha_confirmacion_pickup')
```
